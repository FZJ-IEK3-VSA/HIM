## Functions

```python
def import_shp(path, crs={'init' :'epsg:4326'}, name=""):
    gpdDataFrame=gpd.read_file(path)
    gpdDataFrame=gpdDataFrame.to_crs(crs)
    gpdDataFrame["intIndex"]=gpdDataFrame.index.values
    gpdDataFrame.index=[name+str(id1) for id1 in gpdDataFrame.index.values]
    gpdDataFrame.ID=gpdDataFrame.index.values
    return gpdDataFrame
```

For the calculation of the pipeline routing, we need the 3 differente routing options as one graph.
First, we add the three DataFrames to one DataFrame.

```python
dfLines=dfHighway.geometry.append(dfNaturalGas.geometry).append(dfRailway.geometry)
```

Secondly, we unify the DataFrame to one LineString and split it up afterwards into single lines. IN case we have MultiLineStrings (Lines with curves) we need to get deeper and split these points too.

```python
lines=[line for line in dfLines]
union=gpd.GeoSeries(shp.ops.cascaded_union(lines))
dfLinesSingle=gpd.GeoSeries([line for line in union[0]])
dfLinesSingle=gpd.GeoDataFrame(dfLinesSingle, columns=["geometry"])
```

```python
lines=[]
pIn=[]
pOut=[]
for line in dfLinesSingle.geometry:
    for ix in range(len(line.coords)-1):
        start=line.coords[ix]
        pIn.append(start)
        stop=line.coords[ix+1]
        pOut.append(stop)
        lines.append(LineString([Point(start),Point(stop)]))
```

```python
dfLinesFinal=gpd.GeoDataFrame(lines, columns=["geometry"])
dfLinesFinal["coordsIn"]=pIn
dfLinesFinal["coordsOut"]=pOut
dfLinesFinal["indexIn"]=0
dfLinesFinal["indexOut"]=0
```

# Hydrogen centroid manipulation
The centroids of each state have to be elaborated and have to be in the right coordinate system (Import from Laras data)


Option2: get centroids via District.geometry.centroids()

```python
CentroidNew=gpd.GeoDataFrame.from_dict({'Bayern': (11.4456041081, 48.8622333666),
                                        'Niedersachsen': (9.12722223918, 52.6858662381),
                                        'Mecklenburg-Vorpommern': (12.4698556558, 53.8146418068),
                                        'Baden-Württemberg': (8.89322923406, 48.6411938688),
                                        'Nordrhein-Westfalen': (7.1639949346, 51.3944295551), 
                                        'Schleswig-Holstein': (9.67024308685, 53.9400550439), 
                                        'Sachsen-Anhalt': (12.0947866882, 51.7270829916), 
                                        'Thüringen': (11.0473720853, 50.9180259376), 
                                        'Sachsen': (13.2312279138, 51.0315042369), 
                                        'Rheinland-Pfalz': (7.6625614833, 49.7028466017), 
                                        'Hessen': (8.90655237355, 50.494522903), 
                                        'Brandenburg': (13.7149285363, 52.5539317685)}).T
CentroidNew.geometry=[Point((CentroidNew[0][ix], CentroidNew[1][ix])) for ix in CentroidNew.index]
CentroidNew.crs={'init' :'epsg:4326'}
CentroidNew=CentroidNew.to_crs({'init' :'epsg:7416'})
```

To find the shortest route to add the Centroid to the potential pipeline route, we have to find the nearest point of the route for each centroid and add a Line between the centroid and the nearest point. As Preprocessing for the plotting, we store the Coordinates in x and y too.

```python
nearestPoint=[]
MPoints=MultiPoint(dfLinesFinal["coordsIn"])
for ix in CentroidNew.index:
    (Cent, nP)=shp.ops.nearest_points(CentroidNew.geometry[ix], MPoints)
    nearestPoint.append(nP.coords[0])
```

```python
CentroidNew["nPPoints"]=[Point(nP) for nP in nearestPoint]
CentroidNew["nPCoords"]=nearestPoint
CentroidNew["coords"]=CentroidNew.geometry.map(lambda g: (g.x, g.y))
```

```python
CentroidLine=gpd.GeoDataFrame([LineString([CentroidNew.geometry[ix], CentroidNew["nPPoints"][ix]]) for ix in CentroidNew.index])
CentroidLine["geometry"]=CentroidLine[0]
```

Next to the position of the centroids, we need the incidence matrix to elaborate potential connections between two districts

```python
incidences=[{'Bayern', 'Baden-Württemberg'}, {'Hessen', 'Baden-Württemberg'}, {'Rheinland-Pfalz', 'Baden-Württemberg'}, {'Hessen', 'Bayern'}, {'Sachsen', 'Bayern'}, {'Thüringen', 'Bayern'}, {'Mecklenburg-Vorpommern', 'Brandenburg'}, {'Brandenburg', 'Niedersachsen'}, {'Brandenburg', 'Sachsen-Anhalt'}, {'Sachsen', 'Brandenburg'}, {'Hessen', 'Niedersachsen'}, {'Nordrhein-Westfalen', 'Hessen'}, {'Rheinland-Pfalz', 'Hessen'}, {'Thüringen', 'Hessen'}, {'Mecklenburg-Vorpommern', 'Niedersachsen'}, {'Mecklenburg-Vorpommern', 'Schleswig-Holstein'}, {'Nordrhein-Westfalen', 'Niedersachsen'}, {'Sachsen-Anhalt', 'Niedersachsen'}, {'Schleswig-Holstein', 'Niedersachsen'}, {'Thüringen', 'Niedersachsen'}, {'Nordrhein-Westfalen', 'Rheinland-Pfalz'}, {'Sachsen', 'Sachsen-Anhalt'}, {'Thüringen', 'Sachsen-Anhalt'}, {'Thüringen', 'Sachsen'}]
dfIncidences=pd.DataFrame(incidences)
dfIncidences["start"]=dfIncidences[0]
dfIncidences["target"]=dfIncidences[1]
del dfIncidences[0], dfIncidences[1]
dfIncidences["length"]=[0 for ix in dfIncidences.index]
```

# Graphical calculations with NetworkX
The final pipeline routing is calculated by NetworkX. Therefore, we add 3 empty Graphs (G for everything, drawG as plotting Graph, finalGraph as ResultGraph).

```python
G=nx.Graph()
drawG=nx.Graph()
finalGraph=nx.Graph()
```

We add edges to Graph G as list of (StartNode, TargetNode, lengthOfEdge).

```python
listT=list(zip(dfLinesFinal.coordsIn, dfLinesFinal.coordsOut, dfLinesFinal.length))
listC=list(zip(CentroidNew.coords, CentroidNew.nPCoords, CentroidLine.length))
G.add_weighted_edges_from(listT)
G.add_weighted_edges_from(listC)
```

After preparing the Graph, we can start iterating over the IncidenceMatrix and calculate the shortest path between two centroids and the associated length of the path
