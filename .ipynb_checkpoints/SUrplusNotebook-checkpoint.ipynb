```python
def trunc(value, high, low=0):
    return min(max(value, low), high)

def getFLH(power, args):
    array=args[0]
    FLHTarget=args[1]
    array=array.map(lambda x: trunc(x, power))
#    array[array>power]=power
    FLH=array.sum()/power
    return FLH-FLHTarget

def getFLHReal(power, array):
    array=array.map(lambda x: trunc(x, power))
#    array[array>power]=power
    FLH=array.sum()/power
    return (FLH, array.sum())
```

```python
#%%
electricityDemandRES=50#kWh/kg
#%%
path=os.getcwd() + "\\dataH2Mobility\\"
#surplus=pd.read_csv(path+"surplus_0_BEV.csv", encoding = "ISO-8859-1", index_col="Unnamed: 0")
#plt.style.use(path + "matplotlibrc.mplstyle")
surplus=pd.read_excel(path+"Curtailment_per_node_and_hour.xlsx", index_col="name")
buses=pd.read_excel(path+"kostas_busses.xlsx")
#%%
surplus[surplus<0]=0
surplusRes=pd.DataFrame(columns=surplus.columns)
surplusRes.loc["sum"]=surplus.sum(axis=0)
surplusRes.loc["hydrogenPotential"]=surplusRes.loc["sum"]/electricityDemandRES
surplusRes.loc["P_max"]=surplus.max(axis=0)
surplusRes.loc["P_mean"]=surplusRes.loc["sum"]/8760
surplusRes.loc["percentage"]=surplusRes.loc["P_mean"]/surplusRes.loc["P_max"]
surplusRes.loc["posHours"]=[len(surplus[surplus[county]>0]) for county in surplus.columns]
```

```python
FLHTarget=4400
P_min=10
surplusUsed=surplus.copy()
```

```python
power=[]
for county in surplus.columns:
    if sum(surplus[county])<=0 or surplusRes.loc["posHours", county]<(FLHTarget):
        power.append(0.)
        surplusUsed[county]=0
    else:
        powerOut=(fsolve(getFLH, x0=surplusRes.loc["P_mean",county], args=[surplus[county], FLHTarget])[0])
        if powerOut<P_min:
            powerOut=0
        power.append(powerOut)
        surplusUsed[county]=surplus[county].map(lambda x: trunc(x, powerOut))
    #%%

surplusRes.loc["electrolysisPower"]=power
surplusRes.loc["totalProduction"]=surplusRes.loc["electrolysisPower"]*FLHTarget/electricityDemandRES
res=surplus-surplusUsed
surplusRes.loc["percentualProduction"]=surplusRes.loc["totalProduction"]/surplusRes.loc["hydrogenPotential"]
surplusRelevant=surplusRes.loc[:,surplusRes.loc["electrolysisPower"]>0]
print("total Hydrogen Production in "+ str(len(surplusRelevant.T))+ " counties: " + str(surplusRes.loc["totalProduction"].sum()))
#%%
```
