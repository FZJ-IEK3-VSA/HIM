## Functions

```python
def import_shp(path, crs={'init' :'epsg:4326'}, name=""):
    gpdDataFrame=gpd.read_file(path)
    gpdDataFrame=gpdDataFrame.to_crs(crs)
    gpdDataFrame["intIndex"]=gpdDataFrame.index.values
    gpdDataFrame.index=[name+str(id1) for id1 in gpdDataFrame.index.values]
    gpdDataFrame.ID=gpdDataFrame.index.values
    return gpdDataFrame
```

```python
def NXtoList(NXGraph, attribute):
    dicAttr=nx.get_edge_attributes(NXGraph,attribute)
    return [dicAttr[x] for x in dicAttr], dicAttr
```

```python
def point_array(GeoSeries):
    '''
    create Numpy array from GeoSeries of Points
    input: GeoSeries of points
    '''
    x=GeoSeries.geometry.map(lambda p:p.x).values
    y=GeoSeries.geometry.map(lambda p:p.y).values
    return list(zip(x,y))
```

```python
def distMatrix(gpdIn, gpdOut, kNN=10):
    '''
    Build distance Matrix for input = output system
    Coords: Coordinates as list of tuples from
    Coords2: Coordinates as list of tuples from input
    kNN: number of nearest neighbours
    ID2: array of Coords 2 IDs
    '''
    CoordsOut=point_array(gpdOut)
    CoordsIn=point_array(gpdIn)
    tree=sp.KDTree(CoordsOut,leafsize=3)
    treeDist, treeLoc= tree.query(CoordsIn, k=kNN)
    idx=(gpdIn.intIndex.values+np.zeros((kNN,1),dtype=np.int)).T.ravel()
    inID=np.array([gpdIn.ID[id1] for id1 in idx])
    outID=np.array([gpdOut.ID[id2] for id2 in treeLoc.ravel()])
    
    distMat=pd.DataFrame([inID,
                          outID,
                          treeDist.ravel()/1000],
                         index=["inputID",
                                "targetID",
                                "distance"]).T
    return distMat
```

```python
def selfDistMatrix(gpdDF, kNN):
    '''
    Build distance Matrix for input = output system
    Coords: Coordinates as list of tuples
    kNN: number of nearest neighbours
    ID: array of ID
    '''
    Coords=point_array(gpdDF)
    tree = sp.KDTree(Coords, leafsize=10)
    treeDist, treeLoc = tree.query(Coords, k=kNN+1)
    idx = (gpdDF.intIndex.values+np.zeros((kNN+1,1),dtype=np.int)).T.ravel()    
    inID=np.array([gpdDF.ID[id1] for id1 in idx])
    tarID=np.array([gpdDF.ID[id2] for id2 in treeLoc.ravel()])
    
    distMat = pd.DataFrame([inID,
                            tarID,
                            treeDist.ravel()/1000],
                           index=["inputID",
                                  "targetID",                              
                                  "distance"]).T
    distMat=distMat[distMat.inputID != distMat.targetID]
    
    return distMat
```

```python
def selfDistMatrixFueling(dfGpd, weightFtoF, kNNmax=10):    
    kNN=len(dfGpd)
    if kNN>kNNmax:
        kNN=kNNmax
    Coords=point_array(dfGpd)
    tree = sp.KDTree(Coords, leafsize=10)
    treeDist, treeLoc = tree.query(Coords, k=kNN)
    idx = (np.linspace(start=0, stop=len(dfGpd)-1, num=len(dfGpd),dtype=int)+np.zeros((kNN,1),dtype=np.int)).T.ravel() 
    inID=np.array([dfGpd.index[id1] for id1 in idx])
    tarID=np.array([dfGpd.index[id2] for id2 in treeLoc.ravel()])

    distMat = pd.DataFrame([inID,
                            tarID,
                            treeDist.ravel()/1000],
                           index=["inputID",
                                  "targetID",                              
                                  "distance"]).T
    distMat=distMat[distMat.inputID != distMat.targetID]
    distMat["inputArea"]=dfGpd.ix[distMat["inputID"]]["area"].values
    distMat["targetArea"]=dfGpd.ix[distMat["targetID"]]["area"].values
    distMat["weightID"]=[str(distMat["inputArea"][ix])+"to"+str(distMat["targetArea"][ix]) for ix in distMat.index]
    distMat["weightedLength"]=[weightFtoF[distMat["weightID"][ix]]*distMat["distance"][ix] for ix in distMat.index]
    return distMat
```

```python
def getDiameterSquare(massflow,
                      H2Density=5.7,
                      vPipeTrans=15):
    '''
    massflow: kt per year
    output: diameter in m²
    '''
    
    #=*pi*d^2*H2Density
    ktPerYear_to_kgPerS=1e6/3600/365/24
    d2=massflow*ktPerYear_to_kgPerS*4/(5.7*15*np.pi)
    return d2
```

```python
def getSpecCost(massflow,
                f_grid=1,                
                H2Density=5.7,
                vPipeTrans=15,
                source="Tietze",
                base="throughput"):
    '''
    massflow: massflow in kt per year
    f_grid: Additional factor for weighting results
    H2Density: Density of hydrogen
    vPipeTrans: maximum velocity of hydrogen inside the pipeline

    '''
    diameter=np.sqrt(getDiameterSquare(massflow, H2Density, vPipeTrans))*1000
    if base=="diameter":
        if source=="Krieg":
            A=2.2e-3
            B=0.86
            C=247.5
        else:
            A=2.2e-3
            B=0.86
            C=247.5            
        specCost=(A*diameter**2+B*diameter+C)        
    elif base=="throughput":     
        if source=="Krieg":
            A=474.77
            B=1.3695
        if source=="Tietze":
            A=402.69
            B=0.57099
        else:
            A=474.77
            B=1.3695
        specCost=A*f_grid+B*massflow    
    
    elif base=="diameter2":
        if source=="Grueger":
            A=165.179
            B=1339.286
        else:
            A=165.179
            B=1339.286
        specCost=(A*f_grid+B*getDiameterSquare(massflow, H2Density, vPipeTrans))
        
    
    return specCost*1e-6
```

```python
def getSpecCostKrieg(massflow,
                f_grid=1,                
                H2Density=5.7,
                vPipeTrans=15,
                ):
    '''
    massflow: massflow in t per year
    f_grid: Additional factor for weighting results
    H2Density: Density of hydrogen
    vPipeTrans: maximum velocity of hydrogen inside the pipeline
    '''
    A=2.2e-3
    B=0.86
    C=247.5

    diameter=np.sqrt(getDiameterSquare(massflow, H2Density, vPipeTrans))*1000
    return (A*diameter**2+B*diameter+C)
```

```python
def getEdgeAttributeFromNX(Graph, attribute):
    '''
    Function fordirect conversion of NetworkX Attributes into an list
    '''
    dicAttr=nx.get_edge_attributes(Graph, attribute)
    return [dicAttr[i] for i in dicAttr]
toc=time.time()
print(str(toc-tic)+" Seconds")
```

```python
G_weighted=nx.Graph()
G_weighted.add_weighted_edges_from(list(zip(distStoG.inputID, distStoG.targetID, distStoG.distance)),
                                   f_weight=weightStoG,
                                   capacity=0,
                                   lineCost=0,
                                   lineCostWeight=0)
G_weighted.add_weighted_edges_from(list(zip(distStoC.inputID,distStoC.targetID,distStoC.distance)),
                                   f_weight=weightStoC,
                                   capacity=0,
                                   lineCost=0,
                                   lineCostWeight=0)
G_weighted.add_weighted_edges_from(list(zip(distCtoC.inputID, distCtoC.targetID, distCtoC.distance)),
                                   f_weight=weightCtoC,
                                   capacity=0,
                                   lineCost=0,
                                   lineCostWeight=0)
G_weighted.add_weighted_edges_from(list(zip(distCtoG.inputID, distCtoG.targetID, distCtoG.distance)),
                                   f_weight=weightCtoG,
                                   capacity=0,
                                   lineCost=0,
                                   lineCostWeight=0)
G_weighted.add_weighted_edges_from(list(zip(distGtoG.inputID, distGtoG.targetID, distGtoG.distance)),
                                   f_weight=weightGtoG,
                                   capacity=0,
                                   lineCost=0,
                                   lineCostWeight=0)
```

### Algorhythm --> Dijkstra shortest path

```python
x=0
y=len(distCtoS_final.index)
tic=time.time()

for inputID in distCtoS_final.index:
    f_grid=1#x/y
    targetID=distCtoS_final.targetID.ix[inputID]
    H2Demand=distCtoS_final.H2Demand.ix[inputID]

    for key in weightEdge:
        iterEdge=G_weighted[key[0]][key[1]]
        iterEdge["addCost"]=iterEdge['f_weight']\
                            *getSpecCost((iterEdge["capacity"]+H2Demand), f_grid=f_grid)\
                            *iterEdge['weight']\
                            -iterEdge["lineCostWeight"]
        G_weighted[key[0]][key[1]]=iterEdge

    path=nx.shortest_path(G_weighted, source=inputID, target=targetID, weight= "addCost")
    lengthP=nx.shortest_path_length(G_weighted, source=inputID, target=targetID, weight= "addCost")
    P=nx.Graph()
    P.add_path(path, addCost=0)
    newEdgesToWeight=nx.get_edge_attributes(P, "addCost")

    for key in P.edges():    
        iterEdge=G_weighted[key[0]][key[1]]
        iterEdge["capacity"]=iterEdge['capacity']+H2Demand
        iterEdge["lineCostWeight"]=iterEdge['f_weight']*getSpecCost(iterEdge["capacity"], f_grid=f_grid)*iterEdge['weight']
        iterEdge["lineCost"]=getSpecCost(iterEdge["capacity"])*iterEdge['weight']
        G_weighted[key[0]][key[1]]=iterEdge   
    
    x+=1
toc=time.time()
```

```python
finalNetwork=nx.Graph()
totalEdgesWeight=nx.get_edge_attributes(G_weighted, "f_weight")

for key in totalEdgesWeight:
    if G_weighted[key[0]][key[1]]['capacity'] >0:
        if G_weighted[key[0]][key[1]]["capacity"]<minimumCapacity:
            capacity=minimumCapacity
        else:
            capacity=G_weighted[key[0]][key[1]]["capacity"]
        length=G_weighted[key[0]][key[1]]["weight"]
        finalNetwork.add_edge(key[0],
                              key[1],
                              length=length,
                              capacity=capacity,
                              diameter=sqrt(getDiameterSquare(G_weighted[key[0]][key[1]]["capacity"]))*1000,
                              lineCost=getSpecCost(capacity)*length,
                              lineCostKrieg=getSpecCost(capacity, base="diameter")*length,
                              lineCostTietze=getSpecCost(capacity, base="diameter", source="Tietze")*length)
        finalNetwork.add_node(key[0], pos=CoordsTotal.coordinates[key[0]])
        finalNetwork.add_node(key[1], pos=CoordsTotal.coordinates[key[1]])
```

```python
Centroid["minFS"]=[round(Centroid["H2Demand_kt"][ix]/fuelingMax_kt_a+0.5) for ix in Centroid.index]
Centroid["realFS"]=[len(Fueling[Fueling["ID_C"]==ix]) for ix in Centroid.index]
Centroid["highwayFS"]=[sum(Fueling[Fueling["ID_C"]==ix]["BAB"]) for ix in Centroid.index]
Centroid["minFS"]=Centroid["minFS"]#+Centroid["highwayFS"]
```

## Dijkstra like Transmission

```python
x=0
y=len(distCtoS_final.index)
tic=time.time()

for inputID in FuelingNew.index:
    f_grid=1#x/y
    targetID=FuelingNew["ID_C"].ix[inputID]
    H2Demand=Centroid["H2Demand_kt_F"].ix[targetID]

    path=nx.shortest_path(G_dist, source=inputID, target=targetID, weight= "weight")
    lengthP=nx.shortest_path_length(G_dist, source=inputID, target=targetID, weight= "weight")
    P=nx.Graph()
    P.add_path(path, addCost=0)
    newEdgesToWeight=nx.get_edge_attributes(P, "addCost")

    for key in P.edges():    
        iterEdge=G_dist[key[0]][key[1]]
        iterEdge["capacity"]=iterEdge['capacity']+H2Demand
        G_dist[key[0]][key[1]]=iterEdge   
    x+=1
toc=time.time()
print(str(toc-tic) + " Sekunden")
```

```python
tic=time.time()
finalNetworkDist=nx.Graph()
totalEdgesWeight=nx.get_edge_attributes(G_dist, "weight")

for key in totalEdgesWeight:
    if G_dist[key[0]][key[1]]['capacity'] >0:
        if G_dist[key[0]][key[1]]['capacity'] <minimumCapacity:
            capacity=minimumCapacity
        else:
            capacity=G_dist[key[0]][key[1]]["capacity"]
        length=G_dist[key[0]][key[1]]["weight"]
        finalNetworkDist.add_edge(key[0],
                                  key[1],
                                  length=length,
                                  capacity=capacity,
                                  diameter=sqrt(getDiameterSquare(capacity))*1000,
                                  lineCost=getSpecCost(capacity)*length,
                                  lineCostKrieg=getSpecCost(capacity, base="diameter")*length,
                                  lineCostTietze=getSpecCost(capacity, base="diameter", source="Tietze")*length)
        finalNetworkDist.add_node(key[0], pos=CoordsTotal.coordinates[key[0]])
        finalNetworkDist.add_node(key[1], pos=CoordsTotal.coordinates[key[1]])
toc=time.time()
print(str(toc-tic) + " Sekunden")
```
