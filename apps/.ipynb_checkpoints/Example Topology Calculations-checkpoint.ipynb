# Workflow for Examplary Topology Calculations

In this application, we showcase the capability for pre-process network graphs and simplify their topologies

The workflow is structured as follows:
1. Import relevant modules / tools / packages
2. Demand and Supply Distribution
3. Creation of Distance Matrices
4. Graph Setup    
    4.1 Basic Graph with Full Nodes    
    4.2 Graph with Reduced Nodes    
    4.3 Minimum Spanning Tree Graph
5. Postprocessing

##  1. Import relevant modules / tools / packages

```python
from HIM.workflow.scenarioExample import *
```

```python
from HIM import dataHandling as sFun
from HIM import optiSetup as optiFun
from HIM import hscTotal
from HIM import plotFunctions as pFun
from HIM.workflow import preprocFunc as preFun   
```

```python
dataPathExample=path.join(dataPath,"TopologyExampleInputs")
```

```python
pathGasL = path.join(dataPathExample,"ExampleGrid")
pathDist= path.join(dataPathExample,"area")
pathSource = path.join(dataPathExample,"randomSources")
```

## Import Georeference Data

```python
gpdLines=sFun.import_shp(pathGasL, crs=crs)
area=sFun.import_shp(pathDist, crs=crs, name="C")
Source=sFun.import_shp(pathSource, crs=crs, name="S")
```

```python
def plot_shape(figsize=(7,10), facecolor="white"):
    unary=gpd.GeoSeries([area.unary_union])
    ax=unary.plot(facecolor=facecolor, figsize=figsize)
    ax.xaxis.set_ticklabels(["" for x in ax.xaxis.get_ticklabels()])
    ax.yaxis.set_ticklabels(["" for x in ax.yaxis.get_ticklabels()])
    ax.axis("off")
    ax.set_aspect("equal")
    return ax
```

The geopandas Dataframe has now linestrings which have long distances and no crossings implemented.

```python
sFun.simple_gpd(gpdLines)
gpdPoints=sFun.rem_dupl_un(gpdLines)
gpdPoints.index=gpdPoints["coords"]
gpdLines["inputID"]=gpdPoints.loc[gpdLines["coordsIn"].values, "ID"].values
gpdLines["targetID"]=gpdPoints.loc[gpdLines["coordsOut"].values, "ID"].values
gpdLines=gpdLines.loc[[values["inputID"]!=values["targetID"] for key, values in gpdLines.iterrows()]].copy()
gpdLines["ID"]=[(values["inputID"],values["targetID"]) for key, values in gpdLines.iterrows()]
gpdLines["distance"]=gpdLines.length/1000
gpdLines["weightedDistance"]=gpdLines["distance"]
```

```python
cmap="nipy_spectral"
```

```python
gpdLines["color"]=[np.random.rand() for x in gpdLines.index]
ax=plot_shape()
gpdLines.plot(ax=ax, cmap=cmap, column="color")
#gpdPoints.plot(ax=ax, color="black")
plt.show()
```

# 2. Demand and Supply Distribution

```python
Centroid=area.copy()
Centroid.geometry=Centroid.centroid
Centroid["FCEV"]=Centroid["vehicles"]
Centroid["H2Demand_kt"]=Centroid["FCEV"]*specificDemand*mileage*1e-6
area["H2Demand_kt"]=Centroid["H2Demand_kt"].values
```

To fasten up the calculations, we slice the Demand to the first 10 Centroids

```python
#Centroid=Centroid.iloc[:50]
```

```python
totalH2Demand=Centroid["H2Demand_kt"].sum()
```

```python
Source["H2ProdCap_kt"]=totalH2Demand/len(Source.index)
```

# 3. Creation of Distance Matrices

```python
distStoG=sFun.distMatrix(Source, gpdPoints, kNN=1, weight=1.4)
distCtoG=sFun.distMatrix(Centroid, gpdPoints, kNN=1, weight=1.4)
distStoC=sFun.distMatrix(Source, Centroid, kNN=1, weight=1.4)
distCtoC=sFun.selfDistMatrix(Centroid,kNN=3, weight=1.4)
```

```python
distGtoG=gpdLines.loc[:,distCtoC.columns]
distGtoG.index=[(values["inputID"], values["targetID"]) for key, values in distGtoG.iterrows()]
```

```python
distMat=distStoG.append(distCtoG).append(distCtoC).append(distGtoG).append(distStoC)
```

```python
gpdPoints.index=gpdPoints["ID"]
```

```python
Locations=Source.geometry.append(Centroid.geometry).append(gpdPoints.geometry)
```

# 4. Graph Setup

```python
timeLimit=60
```

## 4.1 Basic Graph with full nodes

```python
GraphBase=optiFun.PipeNetWork()
GraphBase.initializeEdges(distMat)
```

```python
GraphBase.number_of_edges()
```

```python
GraphBase.number_of_nodes()
```

```python
nx.set_node_attributes(GraphBase, "productionMax", Source.H2ProdCap_kt.to_dict())
nx.set_node_attributes(GraphBase, "demand", Centroid["H2Demand_kt"].to_dict())
```

```python
graphresults1=sFun.getGpdFromNXEdges(GraphBase, Locations, minCapacity=-1)
```

```python
ax=plot_shape()
graphresults1.plot(ax=ax)
plt.show()
```

```python
GraphB=GraphBase.copy()
```

```python
GraphB.initOpti()
GraphB.optModel(timeLimit=timeLimit)
resultsBaseMILP=GraphB.getEdgesAsGpd(Locations, "pipeline")
ax=plot_shape()
resultsBaseMILP.plot(ax=ax, linewidth=resultsBaseMILP["diameter"]*10)
plt.show()
```

```python
GraphB.initOpti(linear=True)
GraphB.optModel(timeLimit=30)
resultsBaseLP=GraphB.getEdgesAsGpd(Locations, "pipeline")
ax=plot_shape()
resultsBaseLP.plot(ax=ax, linewidth=resultsBaseLP["diameter"]*10)
plt.show()
```

## 4.2 Basic Graph with reduced nodes

```python
Graph=GraphBase.copy()
Graph.reduceNetworkSize()
```

```python
graphresults2=sFun.getGpdFromNXEdges(Graph, Locations, minCapacity=-1)
```

```python
ax=plot_shape()
graphresults2.plot(ax=ax)
plt.show()
```

```python
Graph.initOpti()
Graph.optModel(timeLimit=timeLimit)
resultsMILP=Graph.getEdgesAsGpd(Locations, "pipeline")
ax=plot_shape()
resultsMILP.plot(ax=ax, linewidth=resultsMILP["diameter"]*10)
plt.show()
```

```python
Graph.initOpti(linear=True)
Graph.optModel()
resultsLP=Graph.getEdgesAsGpd(Locations, "pipeline")
ax=plot_shape()
resultsLP.plot(ax=ax, linewidth=resultsLP["diameter"]*10)
plt.show()
```

## 4.3 Minimum Spanning Tree

```python
GraphMST=GraphBase.copy()
GraphMST.useMinSpanTree()
graphresultsMST=sFun.getGpdFromNXEdges(GraphMST, Locations, minCapacity=-1)
ax=plot_shape()
graphresultsMST.plot(ax=ax)
plt.show()
```

```python
GraphMST.initOpti(linear=True)
GraphMST.optModel()
resultsMST_LP=GraphMST.getEdgesAsGpd(Locations, "pipeline")
ax=plot_shape()
resultsMST_LP.plot(ax=ax, linewidth=resultsMST_LP["diameter"]*10)
plt.show()
```

```python
GraphMST.initOpti(timeLimit=timeLimit)
GraphMST.optModel()
resultsMST_MILP=GraphMST.getEdgesAsGpd(Locations, "pipeline")
ax=plot_shape()
resultsMST_MILP.plot(ax=ax, linewidth=resultsMST_MILP["diameter"]*10)
plt.show()
```

# 5. Postprocessing

```python
resultsNodesEdges=pd.DataFrame()
for graph, name in zip([GraphBase, Graph, GraphMST],
                 ["Base Graph", "Reduced Graph", "Minimum Spanning Tree"]):
    resultsNodesEdges.loc["Number of Nodes", name]=graph.number_of_nodes()
    resultsNodesEdges.loc["Number of Edges", name]=graph.number_of_edges()
```

```python
resultsNodesEdges.T.plot(kind="barh")
plt.show()
```

```python
resultsComparison=pd.DataFrame()
for (graphLP, graphMILP, name) in zip([resultsBaseLP, resultsLP, resultsMST_LP],
                                      [resultsBaseMILP, resultsMILP, resultsMST_MILP],
                                      ["Base Graph", "Reduced Graph", "Minimum Spanning Tree"]):

    resultsComparison.loc["LP", name]=graphLP["lineCost"].sum()

    resultsComparison.loc["MILP"+str(timeLimit)+"s", name]=graphMILP["lineCost"].sum()
```

```python
ax=resultsComparison.multiply(1e-3).T.plot(kind="bar", rot=0, ylim=(0,10))
ax.set_ylabel("pipeline investment [billion EUR]")
plt.show()
```
